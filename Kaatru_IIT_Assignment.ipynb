Goal:
Develop a model to find the variables that are significant in the demand for shared bikes with the available independent variables and report appropriate metrics of your model evaluation.


Plan: Use Lasso Regression model to identify the significant parameters.

```python
#Importing Necessary libraries
import pandas as pd
import numpy as np
```

```python
#Upload the Dataset
dataset = pd.read_csv('day.csv')
dataset
```

```python
dataset.describe()
```

```python
dataset.isnull().sum()
```

Dataset does not have any null values nor any categorical variables except the date-time variable

```python
#Splitting independent and dependent variables
x= dataset.iloc[:,:-1]
y= dataset.iloc[:,-1]  
```

```python
x
```

```python
#Splitting date time variable into three separate columns and removing the original column from x

x['Year'] = x['dteday'].apply(lambda date: date.split('-')[2])

x['Month'] = x['dteday'].apply(lambda date: date.split('-')[1])

x['Day'] = x['dteday'].apply(lambda date: date.split('-')[0])
```

```python
x
```

```python
del x['dteday']
```

```python
x
```

After Preprocessing we are proceeding to Lasso Regression models. Note we have not scaled down the variables in preprocessing. 

```python
#importing necessary libraries
import numpy as np
from sklearn.preprocessing import StandardScaler
from sklearn.pipeline import Pipeline
from sklearn.model_selection import train_test_split, GridSearchCV
from sklearn.linear_model import Lasso
```

```python
#Splitting dataset
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.33, random_state=42)
```

```python
#Applying scaling with Lasso model
pipeline = Pipeline([
                     ('scaler',StandardScaler()),
                     ('model',Lasso())
])
```

```python
#Now we are proceeding to find the alpha variable. Lets try a range of 0.1 to 10
search = GridSearchCV(pipeline,
                      {'model__alpha':np.arange(0.1,10,0.1)},
                      cv = 5, scoring="neg_mean_squared_error",verbose=3
                      )
```

```python
#Fitting the model
search.fit(x_train,y_train)
```

```python
search.best_params_
#We have arrived at the alpha value for the model
```

```python
coefficients = search.best_estimator_.named_steps['model'].coef_
```

```python
#Absolute value
importance = np.abs(coefficients)
```

```python
#All the zeros here mark the columns that have least or no significance
importance
```

```python
features = x.columns
```

```python
#The models that are actually significant.
np.array(features)[importance > 0]
```

Conclusion: 'yr', 'casual', 'registered', 'Year'are the signicant varioables for the dependent variable.
